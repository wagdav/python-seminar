# Charged particle in a homogenous magnetic field

Logic: Hamiltionian -> Symbolic manipulation -> Equations of motion -> Numerical solution


## Hamiltionian

\begin{equation}
H = \frac{1}{2 m} \sum_j (p_j - e  A_j) + q \Phi
\end{equation}
where $\nabla \times \vec{A} = \vec{B}$.

We will use the vector potential
\begin{equation}
\vec{A} = (0, B x, 0)
\end{equation}
that gives a homogenous magnetic field $\vec{B} = B \vec{e}_z$

```python
import sympy

px, py, pz, q, m, B = sympy.symbols('px, py, pz, q, m, B')
qx, qy, qz = sympy.symbols('qx, qy, qz')
```

```python
Ay = B * qx
H = 0.5 * ((px)**2 + (py - q*Ay)**2 + pz**2) / m
```

##Equations of motion

\begin{equation}
q_j = \frac{\partial H}{\partial p_j} \qquad p_j = - \frac{\partial H}{\partial q_j}
\end{equation}

```python
d = sympy.diff
qdot = [d(H, px), d(H, py), d(H, pz)]
pdot = [-d(H, qx), -d(H, qy), -d(H, qz)]
```

## Numerical solution

We write the system in the form of

$\frac{\mathrm{d}y}{\mathrm{d}t} = rhs(y, t, B, m, e)$

where

$y = (q, p)^{T}$

```python
def rhs(y, t, B, mass, charge):
    """Right-hand side of the Hamiltionian equations."""
    q = y[:3]
    p = y[3:]
    # map the symbols to function arguments
    s = {'qx': q[0], 'qy': q[1], 'qz': q[2],
         'px': p[0], 'py': p[1], 'pz': p[2],
         'B': B, 'm': mass, 'q': charge}
    return [exp.evalf(subs=s) for exp in (qdot + pdot)]
```

### Initial condition

```python
import scipy.constants as const

B0 = 1.44
mass = const.atomic_mass
charge = const.elementary_charge

vperp = 1e8
vpar = vperp / 2.

T = 2 * pi * mass / (charge * B0)
#rL = mass * vperp / (charge * B0)

initial_position = [0, 0, 0]
initial_momentum = [mass * vperp, 0, mass * vpar]
end_time = 2 * T
```

```python
from scipy.integrate import odeint

y0 = initial_position + initial_momentum
t = linspace(0, end_time, 100)
y = odeint(rhs, y0, t, args=(B0, mass, charge))

X, Y, Z = y[:, :3].T
```

### Draw the trajectory

```python
import matplotlib.pyplot as plt

f = plt.figure(figsize=(14, 6))
ax = f.add_subplot(1, 2, 1)
ax.plot(X, Y)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_aspect('equal')

ax = f.add_subplot(1, 2, 2)
ax.plot(X, Z)
ax.set_xlabel('$x$')
ax.set_ylabel('$z$')
```
